```python
#função para analise de atracoes permitidas.
#a função retorna uma lista com as atracoes permitidas
def atracoesPermitidas(ano_de_nascimento:int,altura:float) -> list:
    atracoesPermitidas = []
    idade = 2022-ano_de_nascimento

    if altura >= 1.0 and idade >= 3:
        atracoesPermitidas.append('Carrossel')
    if 4 <= idade <= 9 and altura <= 1.3:
        atracoesPermitidas.append('Piscina de bolinhas')
    if altura >= 1.1:
        atracoesPermitidas.append('Montanha-russa')
    
    return atracoesPermitidas
    
#valores de entrada
ano_nascimento = 2016
altura = 1.2

#output
for atracao in atracoesPermitidas(ano_nascimento,altura):
    print(f'O acesso para {atracao} está liberado.')
```

```python
#função para calculo de fatorial
def fatorial(n: int):
    fatorial = 1
    for i in range(1,n+1):
        fatorial *= i 
    return fatorial

#entrada do usuário
n = int(input('Entre com um número para calcular seu fatorial: '))

#pre-processamento de output 
string = str(1)
for i in range(2,n+1):
    string += ' * '+str(i)
resp = '= ' + string + ' =' if n > 1 else '='

#output
print(f'O fatorial de {n} {resp} {fatorial(n)}.')
```

```python
print('Teste de número primo.')

#entrada
numero = int(input('Entre com um número inteiro: '))

#processamento e saída
if numero == 1:
    print(f'O número 1 não é considerado um número primo.')
else:
    divisores = [1]
    for i in range(2,numero):
        if numero%i==0:
            divisores.append(i)
    divisores.append(numero)

    qntd_divisores = len(divisores)
    primo = 'NÃO ' if qntd_divisores > 2 else ''

    string = str(divisores[0])
    for i in range(1,qntd_divisores-1):
        string +=  ', ' + str(divisores[i])
    string += ' e ele mesmo (' + str(divisores[-1]) + ')'
        
    print(f'O número {numero} é divisível por {string}, então {primo}é primo.')
```
