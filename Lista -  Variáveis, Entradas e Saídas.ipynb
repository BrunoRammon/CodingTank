```python
print('Cálculo da posição de um projétil em um dado instante de tempo.')
initial_velocity = float(input('Digite a velocidade inicial do projétil em m/s: '))
initial_position = float(input('Digite a posição inicial do projétil em m: '))
instant_time = float(input('Digite o tempo em que se deseja obter a posição do projétil em s: '))
print(f'A posição do projétil no instante {instant_time} é {initial_position + initial_velocity*instant_time + (-10.0)* instant_time**2.0 /2.0}.')
```

### **Questão 15**
- Dúvida
  - É para apresentar o resultado da hora e da data diretamente ou para formatá-los de alguma forma?

```python
from datetime import datetime
print('Dia e hora atual')
dt = datetime.now()
day = dt.day
month = dt.month
year = dt.year
hour = dt.hour
minute = dt.minute
second = dt.second
print(f'São {hour}H:{minute}M:{second}S do dia {day//10}{day%10}/{month//10}{month%10}/{year}')
```
